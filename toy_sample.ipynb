```python
from utils import load_dataset, preprocessing, TSNE_visualization
import torch
from train import *

dataset = load_dataset('texas')
split_dataset_balanced(dataset, {})
graph = preprocessing(dataset)
```

```python
graph.y
```

```python
model = MLP(in_channels=graph.num_features,
            out_channels=graph.num_class,
            hidden_channels=32,
            num_layers=3,
            dropout=0.5)

device = 'cuda:1'
model.to(device)
model.train()
optimizer = SGD(model.parameters(), lr=0.01, weight_decay=0.001) 
criterion = nn.CrossEntropyLoss()
graph = graph.to(device)
labels = graph.y

best_val_metric = -torch.inf
pred_test = None 

for epoch in range(200):
    optimizer.zero_grad()
    out = model(graph)
    loss = criterion(out[graph.train_index_A], labels[graph.train_index_A])
    
    loss.backward()
    optimizer.step()
    
    metric = cal_accuracy(graph.y[graph.train_index_A], out[graph.train_index_A])
    _, val_metric = eval(model, graph, 'val', 'accuracy')
    print('metric:', metric, 'val_metric: ', val_metric)
    if val_metric > best_val_metric:
        best_epoch = epoch
        best_val_metric = val_metric
        pred_test = torch.softmax(out, dim=1)
pred_test
# metric_record.append(best_val_metric)
# prediction_record.append(pred_test)    
# best_epoch_record.append(best_epoch) 
```
